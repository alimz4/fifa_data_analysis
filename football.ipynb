# Q1

```python
#Max_Overall_liverpool_Barcelona
df10=((df['Club']=='FC Barcelona')| (df['Club']=='Liverpool'))
#df.loc[df10]
df11 = df.loc[df10].sort_values('Overall' , ascending=False)[:5]
df11[['Name' , 'Overall']]
```

# Q2

```python
#RF_Right_Left
df2=(df['Position']=='RF') & (df['Preferred Foot']=='Right')
df3=(df['Position']=='RF') & (df['Preferred Foot']=='Left')
#df.loc[df2]
#df.loc[df3]

right = len(df.loc[df2].index)
left = len(df.loc[df3].index)

percent_right = ((right)/(right + left))*100
percent_left = ((left)/(right+left))*100

print('%right_RF :' , percent_right)
print('%left_RF :' , percent_left)
print(df['Preferred Foot'].value_counts())
```

# Q3

```python
#Nationality_Barcelona
df5 = (df['Club'] == 'FC Barcelona')
df6 = df.loc[df5].sort_values('Age' , ascending=True)[:5]
df6[['Name' , 'Nationality']]
```

# Q4

```python
#Max_Value
plt.figure(figsize=(20,15))

df1=df.sort_values('Value' , ascending=False)[:5]
plt.subplot(2,1,1)
plt.plot(df1['Name'] , df1['Wage'] , color='#5B2C6F',marker='*' ,  markersize=16)
plt.xlabel('Player' , fontsize=20 )
plt.ylabel('Wage' ,  fontsize=20)

plt.subplot(2,1,2)
plt.plot(df1['Name'] , df1['Value'] , marker='*' ,  markersize=16)
plt.xlabel('Player' , fontsize=20 )
plt.ylabel('Value' ,  fontsize=20)

plt.tight_layout()
plt.show()
```

# Q5

```python
#Right_Left
lab = df['Preferred Foot'].value_counts().index
val = df['Preferred Foot'].value_counts().values
explode = [0.2, 0]
plt.pie(val , labels = lab , autopct='%.2f ٪' , explode=explode)
plt.show()
```

# Q6

```python
#Scatter_Overall_Value
plt.figure(figsize=(20,10))

plt.scatter(df['Overall'] , df['Value']  , s=10 , color='b')
plt.xlim([45,95])
plt.xlabel("Overall")
plt.ylabel("Value")
plt.title('Graph', fontsize=20)
plt.xlabel("Overall" , fontsize=20 )
plt.ylabel('Value' ,  fontsize=20)
plt.tight_layout()

plt.show()
#افزایش overall = افزایش Value
```
